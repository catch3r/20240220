```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 50)

y1 = x * 2 + 1
y2 = x ** 2

figure, axes = plt.subplots()

axes.plot(x, y1)
axes.plot(x, y2, color='red', linewidth=1.0, linestyle='--')
axes.set_xlim((-1, 1))
axes.set_ylim((-2, 2))
plt.show()
```

```python
import math

array_x = np.arange(0, 2.5, 0.1)
display(array_x.shape)

y1 = np.sin(math.pi * array_x)
y2 = np.cos(math.pi * array_x)

figure, axes = plt.subplots()
axes.plot(array_x, y1, 'b*')
axes.plot(array_x, y2, 'g^')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = [('apple', 5000), ('banana', 7000), ('orange', 3000), ('tomato', 4000), ('guava', 6000)]
data_df = pd.DataFrame(data, columns=['fruit', 'numbers'])
display(data_df)

figure, axes = plt.subplots()
x_array = data_df['fruit'].values
y_array = data_df['numbers'].values
axes.bar(x_array, y_array)
# display(axes.get_xticks())

# for index, x in enumerate(axes.get_xticks()):
#     print(f'index={index}')
#     print(f'x={x}')

x_ticks_value = axes.get_xticks()
for index in range(len(data)):
    print(x_ticks_value[index], y_array[index])
    axes.annotate(str(y_array[index]), (x_ticks_value[index] - 0.2, y_array[index] + 100))

plt.show()
```
