```python
import matplotlib.pyplot as plt
import numpy as np

index = np.array([3, 4, 5, 6, 7])
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
value1 = [5000, 6000, 7000, 8000, 9000]
value2 = [3000, 4000, 5000, 6000, 7000]
value3 = [4000, 5000, 6000, 7000, 8000]
bin_width = 0.3

figure, axes = plt.subplots()

axes.bar(index, value1, width=bin_width, color='red', label = 'value1')
axes.bar(index + bin_width, value2, width=bin_width, color='blue', label = 'value2')
axes.bar(index - bin_width, value3, width=bin_width, color='green', label = 'value3')

display(axes.get_xticks())
axes.set_xticks(index, labels=subject)

xmin, xmax, ymin, ymax = axes.axis()
xmin, xmax, ymin, ymax

xmin -= 0.2
xmax += 0.2
ymin += 1000
ymax += 1000

axes.axis([xmin, xmax, ymin, ymax])
axes.legend(loc='upper center', ncols=3)
display(axes.get_xticks())
```
