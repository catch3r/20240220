```python
import pandas as pd

data = {
    '縣市': ['台北', '台北', '台北', '新竹', '新竹', '新竹'],
    '年份': [2000, 2001, 2002, 2001, 2002, 2003],
    '人口': [2.2, 2.5, 2.7, 1.1, 1.5, 2.0]
}

df1 = pd.DataFrame(data)
display(df1)
```

```python
#多個相同索引的Series, 可以組成一個DataFrame
display(df1['縣市'])
display(df1['年份'])
display(df1['人口'])
```

```python
#二維的ndArray
import numpy as np

data = np.random.randint(50, 101, size=(50, 5))
scores_df = pd.DataFrame(data, columns=['國文', '英文', '數學', '地理', '歷史'], index=range(1, 51))
scores_df
```

```python
total_series = scores_df['國文'] + scores_df['數學'] + scores_df['英文'] + scores_df['地理'] + scores_df['歷史']
# display(total_series)
# display(scores_df)

scores_df['總分'] = total_series
# display(scores_df)
scores_df['平均'] = total_series / 5.0
display(scores_df)
```
